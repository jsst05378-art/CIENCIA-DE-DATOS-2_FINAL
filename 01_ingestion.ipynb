```python
import fitz  # Importamos PyMuPDF para la lectura del pdf

ruta_pdf = '/content/drive/MyDrive/u/cd2/Banco_Atlas_Financiero.pdf'

def extraer_texto_pdf(ruta):
    texto_completo = ""
    try:
        # Abrimos el documento PDF
        doc = fitz.open(ruta)
        print(f"📄 Documento cargado exitosamente. Total de páginas: {doc.page_count}")

        # Iteramos sobre cada página para extraer el texto
        for num_pagina in range(doc.page_count):
            pagina = doc.load_page(num_pagina)
            # Extraemos el texto y agregamos un salto de línea al final de cada página
            texto_completo += pagina.get_text("text") + "\n"

        return texto_completo

    except Exception as e:
        return f"❌ Error al intentar leer el PDF: {e}"

# Ejecutamos nuestra función
documento_texto = extraer_texto_pdf(ruta_pdf)

# Vamos a imprimir los primeros 1000 caracteres para verificar que se leyó bien
print("\n--- Vista previa del documento ---")
print(documento_texto[:1000])
```

```python
# Guardar el texto en un archivo .txt para usarlo en el siguiente notebook
ruta_guardado = '/content/drive/MyDrive/u/cd2/texto_banco_atlas.txt'

with open(ruta_guardado, 'w', encoding='utf-8') as f:
    f.write(documento_texto)

print("✅ ¡Texto guardado exitosamente en tu Drive!")
```
